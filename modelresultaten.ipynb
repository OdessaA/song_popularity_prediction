```python
import glob

import pandas as pd
from IPython.display import display
```

```python
bestanden = glob.glob("model_resultaten_*.csv")

# Lees alle CSV's in, verwijder lege rijen direct
frames = []
for f in bestanden:
    df = pd.read_csv(f)
    df.dropna(subset=["Model"], inplace=True)  # verwijder rijen zonder model
    df = df[df["Model"].str.strip() != ""]  # verwijder lege strings
    frames.append(df)

resultaten_model = pd.concat(frames, ignore_index=True)

# Sorteer op versie en vaste modelvolgorde
volgorde = [
    "Random Forest",
    "Lineaire Regressie",
    "DummyRegressor (mean)",
    "Ridge",
    "Lasso",
    "ElasticNet",
]
resultaten_model["Model"] = pd.Categorical(
    resultaten_model["Model"], categories=volgorde, ordered=True
)
resultaten_model.sort_values(["Versie", "Model"], inplace=True)

# Toon als tabel
display(
    resultaten_model.style.hide(axis="index").set_caption(
        "Volledig overzicht: prestaties per model en versie"
    )
)
```
